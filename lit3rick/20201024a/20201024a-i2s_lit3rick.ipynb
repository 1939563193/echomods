```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('10k.csv',delimiter=',')
```

```python
plt.figure(figsize=(15, 5))
plt.plot(data[0:256])
```

```python
f = plt.figure(figsize=(15, 5))
ax = f.add_subplot(111)
ax.yaxis.tick_left()
plt.plot(data)

ax = f.add_subplot(111)
ax.yaxis.tick_right()

id_line = [int(k/256) for k in range(len(data))] 
#plt.plot(id_line)
plt.title("All lines for 4k captures (total points = 4000*256)")
plt.show()

```

```python
plt.figure(figsize=(15, 5))

for i in [0,240,256,256*2,255*3,255*7,255*14]:
    plt.plot(data[i*256:i*256+255],label = "Acquisition #"+str(i))
plt.legend()
plt.title("Running i2s capture over 4k lines")
plt.show()

```

```python
plt.figure(figsize=(15, 5))

for i in range(10):
    plt.plot(data[i*254:i*254+255],label = "Acquisition #"+str(i))
i = 253
plt.plot(data[i*254+2:i*254+255+2],label = "Acquisition #"+str(i)+ " - data[253*254+2:253*254+255+2]")
i = 257
plt.plot(data[i*254+2:i*254+255+2],label = "Acquisition #"+str(i)+ " - data[257*254+2:257*254+255+2]")

plt.legend()
plt.title("Running i2s capture over 4k lines - data[i*254:i*254+255]")
plt.show()

```
