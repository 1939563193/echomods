# `20200421a` - comparing acquisitions across platforms

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# /matty/LawA/comparatif/data/20180721a-4.npz
# /matty/LawA/comparatif/data/20191027b-1.npz
# /matty/LawA/comparatif/data/20200112a-1.npz
```

#### Opening the data

All acqs are 128us long at 64Msps. un0rick is 10 bits ADC, lit3rick 12.

```python
t = [x/64.0 for x in range(128*64)]

un0rick_vhdl = np.load("data/20180721a-4.npz")
lit3rick_vhdl = np.load("data/20191027b-1.npz")
lit3rick_verilog = np.load("data/20200112a-1.npz")
lit3rick_verilog["b"],lit3rick_verilog["a"]
```

# Comparing acqs

## Dynamic range of the lit3rick seems adequate (blue graph)

```python
plt.plot(t,lit3rick_vhdl["a"], label='lit3rick_vhdl')
plt.plot(t,lit3rick_verilog["a"], label='lit3rick_verilog')
plt.plot(t,un0rick_vhdl["a"], label='un0rick_vhdl')
plt.legend()
```

## Checking the two graphs without dynamic DAC.. but can't remember which one had which amp 

```python
plt.figure(figsize=(15,5))
plt.plot(t,lit3rick_verilog["a"], label='lit3rick_verilog')
plt.plot(t,un0rick_vhdl["a"], label='un0rick_vhdl')
plt.savefig("signal_check.jpg")
plt.legend()
```

# In this configuration, there's a degraded SNR on the lit3rick board

```python
plt.figure(figsize=(15,5))
plt.plot(t[6000:6500],lit3rick_verilog["a"][6000:6500], label='lit3rick_verilog')
plt.plot(t[6000:6500],un0rick_vhdl["a"][6000:6500], label='un0rick_vhdl')
plt.savefig("noise_check.jpg")
plt.legend()
```
