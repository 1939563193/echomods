```python
import un0usb as USB # neeeds `pip3 install un0usb` before

import matplotlib.pyplot as plt
import numpy as np 

import glob

plt.rcParams["figure.figsize"] = (15,5)

USB.__version__
```

```python
npzs = glob.glob("*.npz")
print(npzs)
USPlot = USB.FView()
```

```python
dataNM = USPlot.readfile('noMatching.npz')
data70 = USPlot.readfile('70pF_GND.npz')
data264 = USPlot.readfile('264pF_GND.npz')
```

```python
plt.plot(dataNM["t"],dataNM["signal"][0],label="No matching")
plt.plot(dataNM["t"],data70["signal"][0],label="70pF")
plt.plot(dataNM["t"],data264["signal"][0],label = "264pF")
plt.xlim(40,45)
plt.ylim(-0.6,0.6)
plt.legend()
plt.savefig("firstEcho.png")
```
