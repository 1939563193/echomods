```python
L = 1000
A = np.arange(L)
```

```python
S = []
E = []
w1 = 0.2*np.pi*0.31
w2 = 2.0*np.pi*0.315
t = 3.0
fig = plt.figure(frameon=False)
for k in range(L):
        tmp = np.exp(-1.0*k*t/L)*(np.sin(w1*k))
        S.append(tmp)
        E.append(1*np.exp(-1.0*k*t/L))
plt.plot(A,S,"y")
plt.plot(A,E,"b")
plt.show()
fig.savefig('RLC2.jpg')
```
