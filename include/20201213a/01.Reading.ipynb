```python
FN = "20201213162644.npz"
```

```python
import un0usb as USB # neeeds `pip3 install un0usb` before
import numpy as np
import matplotlib.pyplot as plt
USB.__version__
```

```python
plot = USB.FView()
data = plot.readfile(FN)
```

```python
np.shape(data["signal"])
```

```python
plt.plot(data["t"],data["signal"][0])
plt.xlim(50,70)
plt.ylim(-0.15,0.15)
```

```python
n = len(data["t"])
f = [x*64.0/n for x in range(n)]
plt.plot(f[100:n//2],np.abs(np.fft.fft(data["signal"][0]))[100:n//2])
```
