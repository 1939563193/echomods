```python
import numpy as np
import matplotlib.pyplot as plt 

import glob, os
os.chdir("./")
S1Ps = []
for file in glob.glob("*.s1p"):
    S1Ps.append(file)
```

```python
for TestName in S1Ps:

    BVP = np.loadtxt(TestName)
    BVP[:,0] = BVP[:,0]/1000000.0

    fig = plt.figure(figsize=(15, 6))
    ax = fig.add_subplot(1, 1, 1)
    x = BVP[:,0]
    y = BVP[:,1]
    z = BVP[:,2]

    Real, = ax.plot(x, y,)
    Imag, = ax.plot(x, z)

    plt.legend(handles = [Real, Imag], 
               labels  = ['Real', 'Imaginary'])
    plt.title(TestName)
    plt.xlabel("Frequency (MHz)")
    plt.show()
    plt.savefig(TestName+"_processed.jpg")
```
