```python
import pyexiv2

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

%matplotlib inline
    

IMG2 = '../../tobo/images/FullSizeRender_1.jpg'
IMG = '../../tobo/images/TEK0006.JPG'
metadata = pyexiv2.ImageMetadata(IMG2)
metadata.read()
```

```python
img = mpimg.imread(IMG2)
plt.imshow(img)
```

```python
metadata.exif_keys
```

```python
tag = metadata['Exif.Image.DateTime']
```

```python
# Writing the description
metadata['Exif.Image.ImageDescription'] = "Original Tobo board - description"
# Original filename
metadata['Exif.Image.DocumentName'] = IMG.split("/")[-1]
# Author
metadata['Exif.Image.Artist'] = "kelu124"
# Related module
metadata['Exif.Image.Software'] = "tobo"
# Type of info -- setup, capture, description, ...
metadata['Exif.Image.ImageHistory'] = "photo description"
# Type of info -- setup, capture, description, ...
metadata['Exif.Photo.MakerNote'] = "Coming from a project aiming at open-sourcing ultrasound imaging hardware - see https://kelu124.gitbooks.io/echomods/content/"
```

```python
metadata.write()
```

```python
for key in metadata.exif_keys:
    print key + " --> "+str(metadata[key].value)
```
